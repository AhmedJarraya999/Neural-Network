```python
import numpy as np
```

```python
def compute_weighted_sum(inputs, weights, bias):
    return np.sum(inputs * weights) + bias
```

```python
def sigmoid(z):
    # Add code here
    #return # Add code here
    return 1 / (1 + np.exp(-z))
```

```python
x1=0.1

#initial weights
w1=0.15
b1=0.4
w2=0.45
b2=0.65

# ground truth
T=0.25

#learning rate
alpha=0.4
```

```python
#define error function MSE
def error(T,a):
    #return # Add code here
     return ((T - a) ** 2).mean()
```

```python
a1= sigmoid(compute_weighted_sum(x1,w1,b1))
print(a1)
```

```python
a2= sigmoid(compute_weighted_sum(a1,w2,b2))
print(a2)
```

```python
E= error(T,a2)
print(E)
```

```python
def dEdw2(a1,a2,T):
    return (a2-T)*a2*(1-a2)*a1
```

```python
def dEdb2(a1,a2,T):
    return (a2-T)*a2*(1-a2)*1
```

```python
def dEdw1(a1,a2,w2,x1,T):
    return (a2 - T) * a2 * (1 - a2) * w2 * a1 * (1 - a1) * x1
```

```python
def dEdb1(a1,a2,w2,x1,T):
      return (a2 - T) * a2 * (1 - a2) * w2 * a1 * (1 - a1)
```

```python
new_w2=w2-alpha*dEdw2(a1,a2,T)
new_b2=b2-alpha*dEdb2(a1,a2,T)
new_w1=w1-alpha*dEdw1(a1,a2,w2,x1,T)
new_b1=b1-alpha*dEdb1(a1,a2,w2,x1,T)


w2=new_w2
b2=new_b2
w1=new_w1
b1=new_b1

print('The updated w2 is {}'.format(np.around(w2, decimals=4)))
print('The updated b2 is {}'.format(np.around(b2, decimals=4)))
print('The updated w1 is {}'.format(np.around(w1, decimals=4)))
print('The updated b1 is {}'.format(np.around(b1, decimals=4)))
```

```python
# what is the new error E?
# let's do forward propagation
a1= sigmoid(compute_weighted_sum(x1,w1,b1))
a2=sigmoid(compute_weighted_sum(a1,w2,b2))
E= error(T,a2)
print('The new error E is {}'.format(np.around(E, decimals=4)))
```
